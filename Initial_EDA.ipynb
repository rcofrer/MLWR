```python
######################################
########### MUTUAL INFO ##############
######################################

def entropy(series):
    """Compute the conditional H(Y|X) in dataframe df
    Args:
        series:   a series from a dataframe
    """
    probs = series.value_counts(dropna=False)/series.count()
    H = -np.sum(probs * np.log2(probs + 1e-100))
    return H 

def cond_entropy(df, Y, X):
    """Compute the conditional H(Y|X) in dataframe df
    Args:
        df:       a dataframe
        X:        the name of the conditioning column
        Y:        the name of the column whose entropy we wish to compute
    """
    vals = df[X].unique()
    ys = []
    for val in vals:
        ent = entropy(df[Y][df[X] == val])
        ys.append(ent)
        
    ys = np.array(ys)
    
    # weights sum up to one (they have to, they are 'probabilities')
    weights = df.groupby(by = X, dropna = False).count()[Y]/len(df)
    
    ################## HARD FIX FOR MUTUAL IFORMATION - HANDLING NaNs ######################
    if len(ys) != len(weights):
        remaining = pd.Series([1 - weights.sum()], index = ["NaN"])
        weights = pd.concat([remaining, weights])
        weights.sort_index()
        np.sort(ys)
        
    cond_H = (ys * weights).sum()
    return cond_H

def mutual_info(df, target, feature, target_entropy):
    """Compute the mututal information in dataframe df between the target and given feature
    Args:
        df:         a dataframe
        target:     the target column
        feature:    the feature column
    """
    return target_entropy - cond_entropy(df, target, feature)

def mutual_info_xy(df,feature1, feature2, entropy_dict):
    
    return entropy_dict[feature1] - cond_entropy(df,feature1,feature2)

# so that we don't need to evaluate it over and over again
has_detections_entropy = entropy(malware['HasDetections'])




############# TO BE UPDATED AFTER #################
########## DEALING WITH THE MISSING DATA ##########
####### MUTUAL INFO CAN'T HAVE MISSING DATA #######

# MI_cols = [col for col in malware.columns if col not in bins and col != 'MachineIdentifier']
MI_cols = [col for col in malware.columns if col not in nums and col != 'MachineIdentifier']
mutual_infos = {}
entropies    = {}
cond_entropies = {}
```
